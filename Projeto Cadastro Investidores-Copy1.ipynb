### 6 - Faça uma análise dos dados faltantes e trate-os¶
#### Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
df5.head(2)
```

```python
# Analisando os dados contando os nulos, somando os valores nulos NaN e mostrando a quantidade de dados nulos em cada coluna.
df5.isnull().sum()
```

```python
#Explicando o que vai se fazer:
#a)-Vamos somar os dados faltantes e colocá-los em ordem decrescente :total = df5.isnull().sum().sort_values(ascending=False)
#b)- Vamos calcular o percentual dos dados faltantes em cada coluna:
#                                                  (df5.isnull().sum()/df5.isnull().count()).sort_values(ascending=False)
#c)- Vamos criar o título das colunas (Total e Percentual): 
```

```python
# verificar missing data (dados faltantes)
total = df5.isnull().sum().sort_values(ascending=False) 
percent = ((df5.isnull().sum()/df5.isnull().count())*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total','Percent'])
missing_data
```

# Tratando os dados Faltantes

```python
#Decidimos preencher os dados faltantes pela média(median)
# Vamos preencher com a média pois configura-se como um valor mais adequado para substituir valores de uma série tão grande,
# pois a utilização da média não representará um desvio muito grande dos valores registrados em cada Ativo.
# Por isso a opção foi pela média que substituirá os valores faltante com a média de cada coluna

#Preenchendo os ativos com a média(mean)
df5['Ativo01'].fillna(df5['Ativo01'].mean(), inplace=True) 
df5['Ativo02'].fillna(df5['Ativo02'].mean(), inplace=True) 
df5['Ativo03'].fillna(df5['Ativo03'].mean(), inplace=True) 
df5['Ativo04'].fillna(df5['Ativo04'].mean(), inplace=True) 
df5['Ativo05'].fillna(df5['Ativo05'].mean(), inplace=True) 
df5['Ativo06'].fillna(df5['Ativo06'].mean(), inplace=True) 
df5['Ativo01'].fillna(df5['Ativo01'].mean(), inplace=True) 

# Preenchendo as Notas faltantes pela média(mean)
df5['Nota1'].fillna(df5['Nota1'].mean(), inplace=True) 
df5['Nota2'].fillna(df5['Nota2'].mean(), inplace=True)
```
